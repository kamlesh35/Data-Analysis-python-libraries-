# Distplots
* Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.

*Import Matplotlib,Import Seaborn*

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0,1,2,3,4,5])
plt.show()
```

* this graph is called bell graph

*Ploting a distplot without the histogram*


```python
sns.distplot([0,1,3,4,5,6],hist=False)
plt.show()
```

# Normal (Gaussion) Distribution
*Normal Distribution*
* The Normal Distribution is one of the most important distributions.
* It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.
* It fits the probability distribution of many events, eg. IQ Scores, Heartbeat etc.
* Use the random.normal() method to get a Normal Data Distribution.
* It has three parameters:
* loc - (Mean) where the peak of the bell exists.
* scale - (Standard Deviation) how flat the graph distribution should be.
* size - The shape of the returned array.

```python
# Generate a random normal distribution of size 3X3 with mean at 1 and standard 
# Deviation of 2
from numpy import random
x = random.normal(loc=1,scale = 2,size = (3,3))
print(x)
```

# Visualization of normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=1000),hist=False)
plt.show()
# Note: The curve of a Normal Distribution is also known as the Bell
#  Curve because of the bell-shaped curve.

```

```python
sns.displot(random.normal(size=1000))
plt.show()
```

# Binomial Distribution 
* Binomial Distribtion is a discreate distribution 
* It discribe the outcome of binary scenarios.
* It has three parameter
* 1- Number of trail
* 2- Probilty of Occurance of each trails
* size - the shape of return array

```python
x = random.binomial(n=10,p=0.1,size=10)
print(x)
```

```python
sns.distplot(random.binomial(n=10,p=0.5,size=1000),hist=True,kde=False)
plt.show()
```

# Diffrence Between Normal and Binomial Distribution
* The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.

```python
sns.histplot(random.normal(loc=50,scale=5,size=1000) ,label='normal')
sns.histplot(random.binomial(n=100,p = 0.5,size=1000),label='Binomial')
plt.show()
```

# Poisson Distribution 
*Poisson Distribution*
* it estimates how many times an event can happen in a specific time 
*There have tow parameter*
* 1- lam - rate of known number of occurance e.g 2 foe above problem
* 2- size - The shape of the return array

```python
# Generate a random 1X10 distribution of occurance 2:
from numpy import random
x = random.poisson(lam=2,size=10)
print(x)
```

*Visualization of poisson Distribution* 

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=2,size=1000),kde=False)
plt.show()
```
