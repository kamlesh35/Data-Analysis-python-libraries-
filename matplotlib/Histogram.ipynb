# Matplotlib Histogram
* A histogram is a graph showing frequency distribution .
* it is a graph showing the number of observation within each given interval.
* eg - say you ask for the height of 250 people ,you might end up with a hostogram 
# Create Histogram
* use the hist() function to create a histogram.
* the hist() function will use an array of number to create a histogram,the array is sent into a function as an argument.


```python
#  A normal data Distribution by Numpy
import numpy as np
x = np.random.normal(170,20,155)
print(x)
```

*The hist() function will read the array and produce a histogram*

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(170,20,166)
plt.hist(x)
plt.show
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np 
import figure
```

```python
python = np.random.randint(10,22,40)
java = np.random.randint(50,70,25)
php = np.random.randint(30,80,50)
plt.figure(figsize=(15,9))
plt.hist([python,java,php],label =["python","java","php"],rwidth=5,color=['r','m','g'])
plt.suptitle("AllDATA")
plt.title("DATA")
plt.xlabel("XDATA")
plt.ylabel("YDATA")
plt.legend()
plt.show()

```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(16,8))
plt.hist([python,java,php],histtype="bar",color=["m","y","b"],rwidth=0.9,orientation="vertical")
plt.show()
```
