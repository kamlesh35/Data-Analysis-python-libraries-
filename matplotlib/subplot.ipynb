# Subplot
*display multiple plot*
* With the subplot() function you can draw multiple plots in one figure:



```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1
x1 = np.array([2,3,4,5])
x2 = np.array([7,5,4,3])
plt.subplot(1,2,1)
plt.plot(x1,x2)

# Plot2
y1 = np.array([3,4,5,3])
y2 = np.array([6,4,3,5])
plt.subplot(1,2,2)
plt.plot(y1,y2)
plt.show()



```

```python
plt.subplot(2,1,1)
plt.plot(x1,x2)

plt.subplot(2,1,2)
plt.plot(y1,y2)
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

# Title in subplot

```python
plt.subplot(1,2,1)
plt.plot(x1,x2)
plt.title("meta")

plt.subplot(1,2,2)
plt.plot(y1,y2)
plt.title("short")
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```
