# Matplotlib Bar
* With Pyplot, you can use the bar() function to draw bar graphs:



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D','F'])
y = np.array([30,30,20,60,70])
plt.bar(x,y)
```

* The Bar() Fuucntion takes arguments that discribes the layout of the bar.
* The categories and their value represented  by the first and second argument as array.

```python
x = np.array(['A','B','C','D','E'])
y = np.array([100,400,290,300,600])
plt.bar(x,y)
```

# Horizonal Bar
* If you want to the barsto be displayed horizontally instead of vertically use the barh() function :

```python
plt.barh(x,y)
```

*Bar Color*
* COLOR NAME You can use any of the 140 supported color names. eg- hotpink
* COLORHEX - Or you can use Hexadecimal color values: - eg -  #"4CAF50"




```python
plt.bar(x,y,color = 'red')
```

# Bar Width
* The bar() takes the keyword argument width to set the width of the bars:



```python
plt.bar(x,y,width=0.7)   # Fix the size of the bar
```

```python
plt.barh(x,y,height=0.5)
```
