# What is Matplotlib?
* Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
* Matplotlib was created by John D. Hunter.
* Matplotlib is open source and we can use it freely.
* Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility

```python
import matplotlib.pyplot as plt
```

# Matplotlib ploting
*Ploting-x and ploting-y point*
* The plot() funtion is used to draw point (maker) in a diagram
* by Default,the plot() funtion draws a line from point to point
* the function takes parameter for  specifying point in the diagram.
* parameter 1 is an array containing the point on the x-axis.
* parameter 2 is an array containing the point in the y-axis.
* If we need to plot a line from (1,3) to (8,10) we have to pass tow array (1,8)and (3,10) to the plot function


```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([1,5])
ypoints = np.array([1,9])
plt.plot(xpoints,ypoints,label = "mydata")
plt.plot(xpoints , label = "todaydata")
plt.legend(loc = 1)
plt.show()
```

*Ploting graph without line*

```python
plt.plot(xpoints,ypoints,'o')
plt.show()
```

*Multiple Point*
* Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):



```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

# default x Point
* If we do not specify the point on the x axis ,they will get the default value 0,1,2,3 etc depending on the length of y point.
* So if we take the some example as above ,and leave out of x point 


```python
ypoints = np.array([3,8,1,10,5,7,2,9])
plt.plot(ypoints)   # xpoint generate automatically in ASC 
plt.grid()
plt.show()
```

# Marker 
* we can use the keyword argument *marker*  to emphasize each point with a specific marker   

```python
plt.plot(ypoints,marker = 'o')
plt.show()
```

```python
plt.plot(ypoints,marker = '*')
plt.show()
```

# Format string 
* we can use shortcut string notation parameter to specify the marker
* This parameter is also called fmt and it written with this syntax


```python
plt.plot(ypoints,'go:')
plt.show()

# Line refrence = '-',':','--','-.'
# Color reference = "r","g","b","c","m","y","k","w"
```

*Marker Size*

```python
plt.plot(ypoints,marker = 'o',ms = 14)   # ms (marker size)
plt.show()
```

*Marker Color*

```python
plt.plot(ypoints,marker = 'o',ms = 20,mec = 'g',mfc='r') 
```

# Linestyle
* You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:



```python
plt.plot(ypoints,linestyle = 'dotted')  # Linestyle = "dashed","dotted"
```

# Sorted Syntax
*The line style can be written in a shorter syntax:*
* linestyle can be written as ls.
* dotted can be written as :.
* dashed can be written as --.

```python
plt.plot(ypoints,ls = ':' ,color = 'red',lw = 5)   # lw(Linewidth)
            # change color and size of line 
```

# Labels and title

```python
import numpy as np
import matplotlib.pyplot as plt
hori = np.array([80, 85, 90, 95, 100, 105])
ver = np.array([240, 250, 260, 270, 280, 290])
plt.plot(hori,ver)
plt.axis()
plt.title("Tempreture")
plt.xlabel("xclass")
plt.ylabel("yclass")

plt.show()
```

# Create a title for a plot

```python
plt.plot(hori,ver)
plt.title("Meta data")
plt.xlabel("data")
```

# Set font properties in labels and title

```python
plt.plot(hori,ver)
font1 = {'family':'serif','color':'blue','size':'20'}
font2 = {'family':'serif','color':'green','size':'10'}
plt.title("metashort",fontdict=font1)
plt.xlabel("X-leveldata")
plt.ylabel("y-laveldata")
```

# Position the Title
* You can use the loc parameter in title() to position the title.
* Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'right')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```
