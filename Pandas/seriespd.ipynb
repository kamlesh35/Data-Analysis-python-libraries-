# What is Series

A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

Crate a series and clicing and indexing

```python
A = [1,2,3,4,5,6]
set = pd.Series(A)
set
```

```python
# Clicing
set[4]
```

```python
# Labels in pandas
B = [1,2,3,4,5,6]
my = pd.Series(B, index=['A','B','C','D','E','G'])
my
```

```python
my['G']    # Indexing and Clicing 
```

Key and Value of Object as Series (Dictionary)

```python
val = {'day1':100,'day2':200,'day3':300}  # {'Key':Value}
data = pd.Series(val) 
data
```

Create a series from the data only day1 and day2

```python
d = {         
    'day1':100,
    'day2':200,
    'day3':300
    }
myset = pd.Series(d,index=['day1','day2'])
myset
```

Create a dataframe from two series

```python
set = {
    "SN":[1,2,3],
    "Person":["Kamlesh","pawan","Rohan"]  
}
mydata = pd.DataFrame(set)
mydata
```

# What is Dataframe

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns

```python
set = {
    'colories':[200,300,400],
    'duration':[20,30,40]
}
mydata = pd.DataFrame(set)
mydata
```

```python
# Locate Row 
mydata.loc[2]
```

```python
# Locate A particular limit 
mydata.loc[[0,2]]
```

Name Indexing

```python
set = {
    'colorID':[2,4,5,7],
    'colorname':['red','blue','green','yellow']
}
my = pd.DataFrame(set,index=['Ram','shyam','Guru','Mohan'])
my
```

```python
# Located With the name in index
my.loc[['Mohan']]
```
